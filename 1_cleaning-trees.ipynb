<h3 style="color: #ec300e;">First Impressions</h3>

The dataset is called [**B√§ume (trees)**](https://www.govdata.de/suche/daten/baume) and was downloaded from the **Data Portal for Germany** ([GOVDATA](https://www.govdata.de/) ‚Äì Das Datenportal f√ºr Deutschland) where a lot of open data regarding different topics can be found and downloaded.

- The names of the columns and much information within the dataset are in German (the columns which will be used for the analysis will be **translated to English** right at the beginning of the cleaning process, categories will follow when it is necessary for understanding).
- The data was last updated on **5th December 2024**.
- The dataset contains **13 columns and over 71,000 rows**, each row seems to be **describing one specific tree**.
- Two columns indicate the **location** of the specific tree ("latitude" and "longitude"), four columns contain information regarding the **genus and species** of the tree (German and botanic name), four columns describe **measuring values** of the tree (height, diameter of trunk and crown, circumference of trunk). Lastly, there are two binary coded columns which show whether a tree is an **alley tree** and whether it was adopted/ there is a **sponsorship** for this tree.
- Five columns are of **datatype object** and contain string data, one of these is a **unique identifier** and the other four are **nominal categorial** data. All the other columns are numerical with differences regarding their datatype and their nature: Two columns are **integers**, they are **binary coded nominal categorial** data. The remaining six columns are **float**, two of them are **categorial data** (latitude and longitude) and four **continuous numerical** data describing the measurements of the tree.
- The continuous numerical data is displayed in **different units**: two in meter, two in centimeter.

```python
# storing maximal realistic heights per species in a dictionary
species_height_limits = {
    "tilia platyphyllos": 44.0,
    "populus x canadensis": 38.5,
    "populus nigra": 33.0,
    "acer species": 33.0,
    "quercus robur": 38.5,
    "betula pendula": 33.0,
    "acer pseudoplatanus": 38.5,
    "tilia tomentosa": 33.0,
    "acer rubrum": 27.5,
    "fagus sylvatica": 38.5,
    "aesculus hippocastanum": 27.5,
    "malus domestica": 11.0,
    "quercus species": 33.0,
    "acer ginnala": 11.0,
    "acer platanoides": 27.5,
    "fraxinus excelsior": 38.5,
    "malus sylvestris": 11.0,
    "tilia cordata": 33.0,
    "sorbus intermedia": 16.5
}
```

```python
# function to filter out the data which is most likely faulty
def replace_outliers_height_with_median(group):
    species = group.name
    limit = species_height_limits.get(species, float('inf'))  # no limit if species is not in dictionary
    mask = group > limit
    median_value = group[~mask].median()  # median of valid values
    group[mask] = median_value  # replacing unrealistic values
    return group

# Werte in "height" nach Kategorien in "species" verarbeiten
df2['height'] = df2.groupby('species')['height'].transform(replace_outliers_height_with_median)
```

<h3 style="color: #ec300e;">How to Handle Null Values</h3>

I will be trying to handle null values with giving up as little data as possible:
- As the missing values are all in numerical columns, and the parameters are most likely related to the tree genus/ species, my aim is to **replace the missing values with mean or median of the columns per subgroup (genus or species of tree, tbd)**.
- If this is justifiable will be dependent on the data foundation in each subgroup, therefore, I will be checking first the species and then the genus and the **share of NaNs in each subgroup**.
- Preferably, I would replace the NaNs with median/mean values from **species** because this is the smaller-scaled variable (genus is the superordinate category of species).

<h4 style="color: #bb1010;">Conditional Replacement of Tree Metrics with Median</h4>

<div style="
    padding: 15px;
    margin: 10px 0;
    border: 1px solid #f7b70d;
    border-radius: 4px;
    background-color: #e2ee1e;
    color: #060606;
    font-size: 16px;
    line-height: 1.5;
    word-wrap: break-word;
    text-align: left;">
    <strong>Next Steps:</strong>
<p>If the share of NaNs in that species is <strong>less than 25 %</strong>, I will replace it with the median value of that species.</p>
</div>


```python
# calculating median of height, trunk_circumference and crown_diameter per species
median_values = df3.groupby('species')[['height', 'trunk_circumference', 'crown_diameter']].median()

# function to replace NaNs with the median of the respective category if share of NaNs in that category is not greater than 25 %
def replace_nans_with_median(group):
    nan_ratio = group.isna().mean()  # output of this is share of NaNs in that specific group
    if nan_ratio['height'] <= 0.25:
        group['height'] = group['height'].fillna(median_values.loc[group.name, 'height'])
    if nan_ratio['trunk_circumference'] <= 0.25:
        group['trunk_circumference'] = group['trunk_circumference'].fillna(median_values.loc[group.name, 'trunk_circumference'])
    if nan_ratio['crown_diameter'] <= 0.25:
        group['crown_diameter'] = group['crown_diameter'].fillna(median_values.loc[group.name, 'crown_diameter'])
    return group

# applying function to each category in "species" and keeping the old index
df3 = df3.groupby('species', group_keys=False).apply(replace_nans_with_median)
```
